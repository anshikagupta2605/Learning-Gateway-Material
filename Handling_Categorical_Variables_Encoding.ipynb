```python
data=pd.read_csv('/content/Titanic Data Set CSV File (main).csv',usecols=['Survived','Cabin'])
```

```python
data.head()
```

```python
data['Cabin'].fillna('Missing',inplace=True)
data.head()
```

```python
data['Cabin']=data['Cabin'].astype(str).str[0]
```

```python
data.Cabin.unique()
```

```python
data.groupby(['Cabin'])['Survived'].mean()
```

```python
#grouping the data of cabin on the basis of survived column
ordinal_labels=data.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ordinal_labels2={k:i for i,k in enumerate(ordinal_labels,0)}
```

```python
data['Cabin_ordinal_labels']=data['Cabin'].map(ordinal_labels2)
data.head()
```

```python
#dropping the Cabin column
#data=data.drop(["Cabin"],axis=1)
#data.head()
```

```python
#MEAN ENCODING
```

```python
mean_ordinal=data.groupby(['Cabin'])['Survived'].mean().to_dict()
```

```python
data['mean_ordinal_encode']=data['Cabin'].map(mean_ordinal)
data.head()
```
